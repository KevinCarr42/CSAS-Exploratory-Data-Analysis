```python
import pandas as pd
import matplotlib.pyplot as plt
from helpers import sql

# pandas formatting
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 100)
pd.set_option('max_colwidth', 200)
```

```python
csas_requests = sql("SELECT * FROM csas2_csasrequest")
request_notes = sql("SELECT * FROM csas2_csasrequestnote")
```

```python
deferred_notes = pd.merge(
    csas_requests,
    request_notes[request_notes.note.str.contains('defer', case=False)][['csas_request_id', 'note']],
    left_on='id', 
    right_on='csas_request_id',
    how='right'
)[['id','created_at', 'updated_at', 'status', 'fiscal_year_id', 'advice_needed_by', 'advice_fiscal_year_id', 'note']]

deferred_notes['advice_needed_by'] = deferred_notes.advice_needed_by.dt.year
deferred_notes['created_at'] = pd.to_datetime(deferred_notes.created_at.dt.strftime('%Y-%m-%d'))
deferred_notes['updated_at'] = pd.to_datetime(deferred_notes.updated_at.dt.strftime('%Y-%m-%d'))

deferred_notes.columns = ['id', 'created', 'updated', 'status', 'fy','advice_fy', 'advice_fy_rev', 'note']

if False:
    deferred_notes.to_csv('deferred_notes.csv', index=False)
```
